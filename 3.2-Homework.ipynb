```python
# Create list of 5 mixed elements
firstlist = ["sunshine", 35, 99.5, [5,4,3], "jaunita"]

# Replace part of list

firstlist[0] = "cloudy"
print(firstlist)

# Append item to the list

firstlist.append("coffee, creamer, cruise")
print("new list", firstlist)

#Slice a subsection of the first list, save to different 2nd list

secondlist = (firstlist[0:3])
print("secondlist", secondlist)

# Extend your original list with second list

thirdlist = firstlist + secondlist
print("thirdlist", thirdlist)

#Create new "simlist" containing 5 elements of the same type

simlist = ["pink", "purple", "blue", "green", "yellow"]

# Sort "simlist"

simlist.sort()
print("sorted list", simlist)

# Copy "simlist"

newsimlist = simlist.copy()
print(newsimlist)

# Add 2nd and 3rd list together

fourthlist = secondlist + newsimlist
print("fourthlist", fourthlist)
```

```python
### Tuples

# Create a Tuple with 5 elements

Firsttuple = ("orchid", "peony", "rose", "tulip", "lily")

#Multiply tuple by 3, save to second tuple

Secondtuple = Firsttuple * 3
print("Secondtuple",Secondtuple)

# Sort second tuple

print(sorted(Secondtuple))

#Copy 4 elements from second element to new third tuple

thirdtuple = Secondtuple[2:6]
print("thirdtuple",thirdtuple)

# Add 2nd and 3rd tuple into 5th tuple 

fifthtuple = Secondtuple + thirdtuple
print("fifthtuple", fifthtuple)
 

```

```python
### Elements

#Create a set of 3 elements

firstset = {"jeans", "dress", "slacks"}

# Add a list of fruits

firstset.add("orange, papaya, lemon")
print("firstsetupdate", firstset)

# Add a car element

firstset.add("Tesla")
print("Carfirstset", firstset)

#Create second set with odd items

secondset = {"shoes", "toys", "sugar"}
print("secondset", secondset)

# Save the union of set 1 and 2 to third set

thirdset = firstset.union(secondset)
print("thirdset", thirdset)

# pop an element from 2nd set, print remainder

x = secondset.pop()
print("removed item", x)
print("remainder", secondset)

# Clear the First Set, print result

firstset.clear()
print(firstset)

# Discard and remove elements from 3rd set

thirdset.remove("sugar")
thirdset.discard("dress")

# Print remainder of third set

print("remainder of thirdset", thirdset)




```

```python
### Dictionary

# Create a dictionary with at least 5 vaules of different data types

firstdict = {
    "name": "Travis",
    "store": "Target",
    "brand": "samsung",
    "year": "1983",
    "sport": "boxing"
}

# Print 1 value

print(firstdict["store"])

# replace any value with your name

firstdict["name"] = "Jaunita"
print("changed name", firstdict)

# add your favorite color

firstdict["color"] = "purple"
print("added favorite color", firstdict)

# Add a list to dictionary

firstdict["fruits"] = ("apples, oranges, kiwi")
print("added fruit list", firstdict)

# Print a list of dictionary keys

x = firstdict.keys()
print(x)

# Print a list of dictionary values

y = firstdict.values()
print(y)

# Copy 1st dictionary into a second dictionary

seconddict = firstdict.copy()

print("seconddict", seconddict)

# Pop an item from second dictionary, print

seconddict.pop("sport")
print("seconddict popped sport", seconddict)

# Remove all elements from 2nd dictionary, print result

seconddict.clear()
print("empty seconddict", seconddict)






```
